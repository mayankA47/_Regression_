```python
#multiple linear regression
def y_pred(theta,x):
    y=0
    m=theta.shape[0]
    for i in range(0,m):
        y = y+theta[i]*x[i]
    return y

def gradient(X,Y,theta):
    m=X.shape[0]
    grad = np.zeros(X.shape[1])
    for i in range(0,m):
        y = y_pred(theta,X[i])
        for j in range(0,X.shape[1]):
            grad[j] += (y-Y[i])*X[i][j]
    return grad/m
        
        
    
def multiple_regression(X,Y,theta,lr=0.000001,steps=10000):
    for i in range (0,steps):
        grad = gradient(X,Y,theta)
        theta = theta - lr*grad
    return theta
        
    

    
    
```
